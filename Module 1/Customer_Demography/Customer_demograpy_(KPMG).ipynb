Detecting DOB columns for correcting date of birth value and conver it into corresponding age

```python
#This command "find errors in Date of birth columns"
from datetime import datetime, date
born ='1953-10-12'
print("Born :", born)
born = datetime.strptime(born, "%Y-%m-%d").date()
today = date.today()
print("Age :", today.year - born.year - ((today.month,today.day) < (born.month,born.day)))
```

```python
# This commands "convert date of birth into Current age"
from datetime import datetime, date
for i in cdg['DOB']:
    print(i)
    born = i
    born = datetime.strptime(born, "%Y-%m-%d").date()
    #Get today's date
    today = date.today()
    j = today.year - born.year - ((today.month, today.day) < (born.month, born.day))
    cdg['DOB'] = cdg['DOB'].replace([i],[j])
```

```python
# This command "change column name" (DOB into Age)
cdg.rename(columns = {'DOB': 'Age'}, inplace=True)
cdg
```
